```
# democrats have 73 yes votes on this issue of handicapped infants while republicans have 23 yes votes on this issue which seems significant.
# we have to next do an inferential test to find p value.
```

```
#creating a crosstab table for frequency distribution for issue of handicapped infants
n_crosstab = pd.crosstab(df4.Party, df4.n, normalize='columns')
n_crosstab
```

```
# since we are comparing proportions, we have to apply z test
```

```
import numpy as np
import scipy.stats as st
import scipy.special as sp
# n = number of obervations of yes and no
n = 100  
# h = number of yes votes for democrats on this issue
h = 76  
# q = null-hypothesis of no difference in yes and no votes
q = .5  
```

```
# z score calculation

xbar = float(h) / n
z = (xbar - q) * np.sqrt(n / (q * (1 - q)))
# We don't want to display more than 4 decimals.
z
```

```
pval = 2 * (1 - st.norm.cdf(z))
pval
```

```
#The p value here is very low and lower than 0.01, so our z score calculation was successful in calculating the p value in this case
```

####Democrats support the issue of handicapped infants more than republicans with a p value < 0.01.

```
# from above output, the number of yes votes for this issue of physician fee freeze is 94% for republicans and 5.3% for democrats. 
# Let us now do a z test to find p value.
```

```
# n = number of obervations of yes and no
n2 = 100  
# h = number of yes votes for republicans on this issue
h2 = 94.6  
# q = null-hypothesis of no difference in yes and no votes
q2 = .5  
```

```
# z score calculation

xbar = float(h2) / n2
z2 = (xbar - q2) * np.sqrt(n2 / (q2 * (1 - q2)))
# We don't want to display more than 4 decimals.
z2
```

```
# For issue y=water-project-cost-sharing, there doesn't seem much difference in yes votes in two parties.
```

```
df9 = df3[['Party', 'y']]
df9.head()
```

```
y2_crosstab = pd.crosstab(df9.Party, df9.y, normalize='columns')
y2_crosstab
```

```
# from the above cross tab, yes votes for democrats on this issue are 52.3% and for republicans are 47.6%.
# Let us do a z test to find p value.
```

```
# n = number of obervations of yes and no
n3 = 100  
# h = number of yes votes for democrats on this issue
h3 = 52.3 
# q = null-hypothesis of no difference in yes and no votes
q3 = .5 
```

```
# z score calculation

xbar = float(h3) / n3
z3 = (xbar - q3) * np.sqrt(n3 / (q3 * (1 - q3)))
# We don't want to display more than 4 decimals.
z3
```

```
pval3 = 2 * (1 - st.norm.cdf(z3))
pval3
```

### The p value is 0.64 which is >0.01, so not significant. There is no significant difference in yes votes for republicans and democrats on the issue of water-project-cost-sharing. The p value is also >0.1.
